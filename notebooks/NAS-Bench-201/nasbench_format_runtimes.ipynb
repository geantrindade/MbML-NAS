```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_size = 172
```

```python
df_n172 = pd.read_csv(str("/home/gean/nns_performance_prediction/results/fast/test12/nasbench101_allSubsets_allSeeds_n" + 
                          str(train_size) + "_fit_times.csv"))
df_n172.head(5)
```

```python
df_dict = {'dataset': [], 'epoch': [], 'train_size': [], 'seed': [], 'model': [], 'fit_time': []}

for idx, seed in enumerate([0, 1, 10, 42, 100, 123, 666, 1000, 1234, 12345]):
    for subset in [4, 12, 36, 108]:
        for reg in ['LinearRegression', 'DecisionTreeRegressor', 'SGDRegressor', 'BayesianRidge', 
                    'KNeighborsRegressor', 'SVR', 'AdaBoostRegressor', 'GradientBoostingRegressor', 
                    'RandomForestRegressor', 'DummyRegressor']:
            df_dict['dataset'].append('cifar10')
            df_dict['epoch'].append(subset)
            df_dict['train_size'].append(train_size)
            df_dict['seed'].append(seed)
            df_dict['model'].append(reg)
            df_dict['fit_time'].append(df_n172[reg][idx])
            
df_results = pd.DataFrame.from_dict(df_dict)    
df_results
```

```python
df_dict = {'dataset': [], 'epoch': [], 'train_size': [], 'seed': [], 'model': [], 'fit_time': []}

for subset in [4, 12, 36, 108]:
    df_gcn_n172_c = pd.read_csv(str("/home/gean/nns_performance_prediction/results/gcn/runs_c_train/runs_" + 
                                    str(subset) + ".csv"))
    df_gcn_n172_p = pd.read_csv(str("/home/gean/nns_performance_prediction/results/gcn/runs_p_train/runs_" + 
                                    str(subset) + ".csv"))
    complete_fit_time = list(df_gcn_n172_c['time']) + list(df_gcn_n172_p['time'])
    
    for idx, seed in enumerate([0, 1, 10, 42, 100, 123, 666, 1000, 1234, 12345]):
        df_dict['dataset'].append('cifar10')
        df_dict['epoch'].append(subset)
        df_dict['train_size'].append(train_size)
        df_dict['seed'].append(seed)
        df_dict['model'].append("GCN_Complete")
        df_dict['fit_time'].append(complete_fit_time[idx])

df_results_gcn = pd.DataFrame.from_dict(df_dict)  
df_results = df_results.append(df_results_gcn)                                                                                            
df_results.to_csv(str("/home/gean/nns_performance_prediction/results/fast/test12/" +
                      "nasbench101_cifar10_allEpochs_n" + str(train_size) +
                      "_fit_times.csv"), index=False, float_format='%.6f')
df_results
```
