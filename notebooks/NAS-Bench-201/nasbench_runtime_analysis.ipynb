```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# NAS-Bench-101 (All Subsets/Epochs, N172)

```python
df_n172 = pd.read_csv("/home/gean/nns_performance_prediction/results/fast/test12/nasbench101_cifar10_allEpochs_n172_fit_times.csv")
df_n172
```

```python
df_n172['model'].replace({'LinearRegression': 'LR', 'DecisionTreeRegressor': 'DT', 'SGDRegressor': 'SGD', 
                     'BayesianRidge': 'BR', 'KNeighborsRegressor': 'KNN', 'SVR': 'SVM',
                     'AdaBoostRegressor': 'AB', 'GradientBoostingRegressor': 'GB', 'RandomForestRegressor': 'RF', 
                     'DummyRegressor': 'Dummy', 'GCN_Complete': 'GCN'}, inplace=True)
df_n172.rename(columns={'fit_time': 'Training time (seconds)', 'train_size': 'train size'}, inplace=True)
df_n172
```

```python
sns.set_theme(style="whitegrid")
sns.set(rc={'figure.figsize':(18.7, 13.27)})
sns.set(font_scale=2.5)
```

```python
ax = sns.boxplot(x="model", y="Training time (seconds)", hue="epoch", data=df_n172, palette="colorblind")
ax.set(xlabel=None)
ax.set_title("NAS-Bench-101 (Cifar10), Train size 172")
plt.savefig(str('/home/gean/nns_performance_prediction/results/fast/test12/' + 
                'nasbench101_cifar10_allEpochs_n172_boxplot_fit_times_allModels.pdf'), bbox_inches='tight')
```

```python
ax = sns.boxplot(x="model", y="Training time (seconds)", hue="epoch", data=df_n172.loc[df_n172['model'] != 'GCN'], palette="colorblind")
ax.set(xlabel=None)
ax.set_title("NAS-Bench-101 (Cifar10), Train size 172")
plt.savefig(str('/home/gean/nns_performance_prediction/results/fast/test12/' + 
                'nasbench101_cifar10_allEpochs_n172_boxplot_fit_times_noGCN.pdf'), bbox_inches='tight')
```

```python
ax = sns.boxplot(x="model", y="Training time (seconds)", hue="epoch", data=df_n172.loc[df_n172['model'] == 'GCN'], palette="colorblind")
ax.set(xlabel=None)
ax.set_title("NAS-Bench-101 (Cifar10), Train size 172")
plt.savefig(str('/home/gean/nns_performance_prediction/results/fast/test12/' + 
                'nasbench101_cifar10_allEpochs_n172_boxplot_fit_times_onlyGCN.pdf'), bbox_inches='tight')
```
