## **Usefull methods to extract statistics**

```python
def extract_conv_statistics(conv_layers : list) -> dict:    #conv_layers_info -> int array      
    #None or zero? For now, to avoid bias, I choose zero..  
  conv_statistics = {'conv_num_layers': 0, 'conv_kernel_min': 0, 'conv_kernel_max': 0, 'conv_kernel_mode': 0}

  if (len(conv_layers) > 0):
    conv_statistics['conv_num_layers'] = len(conv_layers)
    conv_statistics['conv_kernel_min'] = min(conv_layers) 
    conv_statistics['conv_kernel_max'] = max(conv_layers) 
    conv_statistics['conv_kernel_mode'] = stats.mode(conv_layers)[0][0]
  
  return conv_statistics
```

```python
def extract_mp_statistics(mp_layers : list) -> dict:
  #conv_layers_info -> int array
  mp_statistics = {'maxpool_usage': 0, 'maxpool_num_layers': 0,}

  if (len(mp_layers) > 0): 
    mp_statistics['maxpool_usage'] = 1
    mp_statistics['maxpool_num_layers'] = len(mp_layers)

  return mp_statistics
```

```python
def get_arch_statistics(arch : list):
  middle_layers = arch[1:-1]

  conv_layers = []
  mp_layers = []

  #for idx, value in enumerate(middle_layers):
  for layer in middle_layers:
    if (layer == 'conv1x1-bn-relu'): conv_layers.append(1)
    elif (layer == 'conv3x3-bn-relu'): conv_layers.append(3)
    elif (layer == 'maxpool3x3'): mp_layers.append(1)

  conv_statistics = extract_conv_statistics(conv_layers)
  mp_statistics = extract_mp_statistics(mp_layers)

  return conv_statistics, mp_statistics

#main
# print(get_arch_statistics(['input', 'maxpool3x3', 'conv1x1-bn-relu', 'maxpool3x3', 'conv1x1-bn-relu', 'conv1x1-bn-relu', 'output']))
```
